```python
def adjacent4_multi(L):
    size = len(L)
    n = int(np.sqrt(size))
    M = 0
    for i in range(3,size):
        if (i-3)%n < n-3:
            across = L[i-3] * L[i-2] * L[i-1] * L[i]
            #print 'Numbers:', L[i-3], L[i-2], L[i-1], L[i],'. Multiplication:', across
            if across > M:
                M = across
                
        if i < size - 3*n:
            right2left = L[i] * L[i+n-1] * L[i+2*(n-1)] * L[i+3*(n-1)]
            #print 'Numbers:', L[i], L[i+n-1], L[i+2*(n-1)], L[i+3*(n-1)],'. Multiplication:', right2left
            if right2left > M:
                M = right2left
                
            left2right = L[i-3] * L[i-3+n+1] * L[i-3+2*(n+1)] * L[i-3+3*(n+1)]
            #print 'Numbers:', L[i-3], L[i-3+n+1], L[i-3+2*(n+1)], L[i-3+3*(n+1)],'. Multiplication:', left2right
            if left2right > M:
                M = left2right
            
            updown = L[i] * L[i+n] * L[i+2*n] * L[i+3*n]
            print 'Numbers:', L[i], L[i+n], L[i+2*n], L[i+3*n],'. Multiplication:', updown
            if updown > M:
                M = updown
                
    return M
```
